```python
import pandas as pd
from login import login_info
from bs4 import BeautifulSoup
import urllib
import time

from selenium import webdriver
from selenium.webdriver.common.keys import Keys

import requests
```

```python
def get_html(driver, y_text, c_text):

    data = driver.page_source
    page = driver.current_url

    soup = BeautifulSoup(data, 'html5lib')  
    table = soup.find('table')

    df = pd.read_html(str(table))[0]
       
    tuples = list(zip([y_text]*len(df), [c_text]*len(df)))

    years = [y_text] * len(df)
    classes = [c_text] * len(df)

    index = pd.MultiIndex.from_tuples(tuples, names=['Years', 'Class'])
    df.index = index

    dfs.append(df)
    
    time.sleep(.05)
    driver.get(url4)
    time.sleep(.05)
```

```python
user, pw = login_info()

url = 'https://my.gavilan.edu:8447/cas-web/login?service=https%3A%2F%2Fmy.gavilan.edu%2Fc%2Fportal%2Flogin'

driver = webdriver.Chrome()
driver.get(url)

ids = driver.find_elements_by_xpath('//*[@id]')

submit = driver.find_elements_by_xpath('//*[@class]')[17]

login_elm, username_elm, password_elm = ids[3], ids[5], ids[6]

username_elm.clear()
time.sleep(.1)
username_elm.send_keys(user)
time.sleep(.1)
password_elm.send_keys(pw)
time.sleep(.1)
submit.click()

url2 = 'https://my.gavilan.edu/web/home-community/intranet'
url3 = 'https://my.gavilan.edu/luminis/gcf/intranet/Authenticate?url=https%3A%2F%2Fmail1.gavilan.edu%2Fslo%2Findex.html'
url4 = 'https://mail1.gavilan.edu/slo/relay.php?mode=instruction'
driver.get(url2)
driver.get(url3)
driver.get(url4)

url5 = 'https://mail1.gavilan.edu/slo/cpslo.php?ay_id={}&cp_id={}&submit=View+Program-Level+Outcomes+%28Instruction%29'

year_options = driver.find_elements_by_xpath("//select[@name='ay_id']")[0]
year_options = year_options.find_elements_by_tag_name("option")[1:]

class_options = driver.find_elements_by_xpath("//select[@name='cp_id']")[0]
class_options = class_options.find_elements_by_tag_name("option")[1:]

year_info = [(y.text, y.get_attribute('value')) for y in year_options]
class_info = [(c.text, c.get_attribute('value')) for c in class_options]

dfs = []

for y_text, y_id in year_info[:]:
    for c_text, c_id in class_info[:]:
        print (y_text, c_text)
        driver.get(url5.format(y_id,c_id))
        get_html(driver, y_text, c_text)

driver.quit()
```

```python
data = pd.concat(dfs)
```

```python
with pd.io.excel.ExcelWriter(path='course.xlsx', engine="xlsxwriter") as writer:
    data.fillna('').to_excel(writer, sheet_name='test')

    workbook = writer.book
    worksheet = writer.sheets['test']
    
    format1 = workbook.add_format({
        'text_wrap':True,
        'align':'center'
    })
    
    format0 = workbook.add_format({'bg_color': 'red'})

    worksheet.set_column('A:A', 15, format1)
    worksheet.set_column('B:F', 30, format1)
    
    abc = workbook.add_format({'bg_color': '#FFC7CE',
                               'font_color': '#9C0006'})
    efg = workbook.add_format({'bg_color': '#C6EFCE',
                               'font_color': '#006100'})
    worksheet.conditional_format(1, 2, 1000, 6,
                             {'type': 'blanks',
                             'format': abc})

    worksheet.conditional_format(1, 2, 1000, 6,
                             {'type': 'no_blanks',
                             'format':    efg})
    writer.save()
```
