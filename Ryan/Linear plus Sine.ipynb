```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline

sns.set()

x = np.arange(0,10,0.1)
x2 = pd.date_range('1/1/2018',periods=len(x))
y1 = np.sin(x*10)

y2 = x +3
y3 = y1 + y2


plt.plot(x2,y1)
plt.plot(x2,y2)
plt.plot(x2,y3)
plt.show()
```
